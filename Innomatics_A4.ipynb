Problem 1: Mobile Recharge Validation System


```python
def validate_recharge():
  valid_plans = [199, 299, 399, 599]
  while True:    #while loop
    try:
      amount = int(input("Enter recharge amount: ₹"))
      if amount <= 50:
        print("Recharge amount must be at least 50. Try again.\n")
      elif amount not in valid_plans:
        print("Invalid Plan selected.")
        print("Available plans are: ", valid_plans)
        print("Please Try again.\n")
      else:
         print("Recharge Successful with ₹", amount)
         break   #exit loop if valid
    except ValueError:
      print("Please enter a valid numeric amount.\n")
validate_recharge()
```

Problem 2: Inventory Reorder Alert System


```python
def check_inventory(stock_data):
  for product, quantity in stock_data.items():  #for loop
    if quantity < 15:
      print(f"{product}: Reorder Alert! (Stock = {quantity})")
    else:
      print(f"{product}: Stock OK (Stock = {quantity})")
inventory = {      #dictionary of products
    "Rice": 25,
    "Wheat": 10,
    "Sugar": 18,
    "Oil":8,
    "Salt":30
}
check_inventory(inventory)
```

 Problem 3: Student Result Processing System

```python
def process_result(marks):
  total = 0
  for mark in marks:   #loop to calculate total marks
    total += mark
  average = total/len(marks)  #calculate average
  if average >= 50:
    return average, "Pass"
  else:
    return average, "Fail"
student_marks = [60, 45, 70, 55]
avg, result = process_result(student_marks)
print("Average Marks:", avg)
print("Result:", result)
```

Problem 4: Cab Fare Estimator with Retry Option


```python
def calculate_fare(distance, is_peak):
  base_fare = 50
  per_km_rate = 12
  fare = base_fare + (distance * per_km_rate)
  if is_peak:
    fare +=fare *0.25  #25% extra
  return fare
while True:   #retry option
  try:
    distance = float(input("Enter distance in km:"))
    peak_input = input("Is it peak hour? (yes/no): ").lower()
    if peak_input =="yes":
      is_peak = True
    else:
      is_peak = False
    total_fare = calculate_fare(distance, is_peak)
    print(f"Total Fare: ₹{total_fare}")
    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
      print("Thank you for using Cab Fare Estimator!")
      break
  except ValueError:
    print("Please enter valid input.\n")
```

Problem 5: Employee Attendance Eligibility Checker

```python
def check_attendance(attendance_list):
  present_count = 0
  for day in attendance_list:   #loop to count present days
    if day == "P":
      present_count += 1
  total_days = len(attendance_list)  #calculate present days
  percentage = (present_count / total_days)* 100
  if percentage >= 75:  #determine eligibility
    return percentage, "Eligible"
  else:
    return percentage, "Not Eligible"
attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]
percent, status = check_attendance(attendance)
print("Attendance Percentage:", percent, "%")
print("Status:", status)
```

Problem 6: Password Strength Checker

```python
def check_password(password):
  has_digit = False
  has_special = False
  special_chars = "@#$"
  if len(password) < 8:
    return "Weak Password(Minimum length should be 8)"
  for char in password:  #loop through each character
    if char.isdigit():
      has_digit = True
    if char in special_chars:
      has_special = True
  if has_digit and has_special: #check conditions
    return "Strong Password "
  else:
    return "Weak Password (Must contain at least one digit and one special character)"
pwd = "PassKey"
result = check_password(pwd)
print(result)
```
