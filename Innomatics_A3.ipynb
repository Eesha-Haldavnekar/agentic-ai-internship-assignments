Problem Statement 1: Employee Performance Bonus Eligibility


```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
highest_score = max(employees.values()) #Finding highest score
top_performers = [name for name, score in employees.items() if score == highest_score] #Finding emp with highest score
print("Top Performers Eligible for bonus:",",".join(top_performers), f"(Score: {highest_score})")
```

Problem Statement 2: Search Query Keyword Analysis


```python
import string
query = "Buy mobile phone buy phone online"
query = query.lower() # Convert to lowercase
query = query.translate(str.maketrans('', '', string.punctuation)) #Remove puctuation
words = query.split() #split into two words
freq = {}
for word in words:
  freq[word] = freq.get(word, 0) +1 #count frequency
result = {word: count for word, count in freq.items() if count > 1}
print(result)
```

Problem Statement 3: Sensor Data Validation

```python
sensor_readings = [3,4,7,8,10,12,5]
valid_readings = []
for index, value in enumerate(sensor_readings):
  if value % 2 == 0: #even no. check
    valid_readings.append((index, value))
print("Valid Sensor Readings(Hour, Value):")
print(valid_readings)
```

Problem Statement 4: Email Domain Usage Analysis


```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
domain_count = {}
for email in emails:
  domain = email.split("@")[1]
  domain_count[domain] = domain_count.get(domain, 0)+1
total = len(emails)
for domain, count in domain_count.items():
  percentage = (count/total) * 100
  print(f"{domain}: {percentage:.0f}%")
```

Problem Statement 5: Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
avg_sales = sum(sales) / len(sales) #Calculate average
print("Days with sales spike (>30% above average):")
for day, value in enumerate(sales, start=1):
  if value > avg_sales * 1.3:
    print(f"Day {day}: {value}")
```

Problem Statement 6: Duplicate User ID Detection

```python
user_ids = ["user1","user2","user1","user3","user1","user3"]
id_count = {}
for user in user_ids:
  id_count[user] = id_count.get(user, 0)+1
for user, count in id_count.items():
  if count > 1:
    print(f"{user} -> {count} times")
```
