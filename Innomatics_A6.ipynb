**Problem Statement 1:Smart Parking Lot Management System**

```python
def smart_parking_sys(capacity, logs):
  current_vehicles = 0
  peak_usage = 0
  for log in logs:     #loop for logs
    if log == "IN":
      current_vehicles += 1
    elif log == "OUT":
      if current_vehicles > 0:   #track peak usage
        current_vehicles -= 1
  print("Currently Parked vehicles:", current_vehicles)
  if current_vehicles > capacity:     #alert for exceeding capacity
    print("Parking Status: Full / OVERCAPACITY ALERT")
  else:
    print("Parking Status: Available")
  print("Peak Parking usage:", peak_usage)

capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
smart_parking_sys(capacity, vehicle_logs)
```

 **Problem Statement 2: Online Food Delivery Time Estimator**

```python
def estimate_del_time(dist, traffic, weather):
  eta = dist * 5
  if traffic =="Low":    #traffic delay
    eta += 5
  elif traffic == "Medium":
    eta += 10
  elif traffic =="High":
    eta += 15

  if weather == "Clear":  #weather delay
    eta += 0
  elif weather == "Cloudy":
    eta += 5
  elif weather == "Rainy":
    eta += 10

  print("Estimated Delivery Time:", eta, "minutes")

estimate_del_time(8, "High", "Rainy")
```

**Problem 3: Movie Theatre Seat Occupancy Analyzer**

```python
def analyze_occupancy(total_seats, booked_seats):
  booked_count = len(booked_seats)
  occupancy = (booked_count / total_seats)*100
  print("Occupancy: ", int(occupancy), "%")
  if occupancy == 100:
    print("Show Status: Housefull")
  elif occupancy >= 75:
    print("Show Status: Almost Full")
  else:
    print("Show Status: Available")

booked = [1] * 150
analyze_occupancy(200, booked)
```

**Problem Statement 4:Cloud Server Load Classification System**

```python
def classify_server_load(cpu_readings):
  total = 0
  for reading in cpu_readings:
    total += reading
  average = total/ len(cpu_readings)
  print("Average CPU load: ", int(average), "%")
  if average < 50:
    print("Server Status: Normal")
  elif average <= 80:
    print("Server Status: Warning")
  else:
    print("Server Status: Critical")

classify_server_load([45, 60, 70, 85, 90])
```

**Problem Statement 5: Smart Classroom Resource Usage Monitor**

```python
def monitor_res(res_usage):
  overused = []
  for res in res_usage:
    if res_usage[res] > 8:
      overused.append(res)
  if overused:
    print("Overused Resources: ", ", ".join(overused))
    print("Energy Alert: Yes")
  else:
    print("Overused Resources: None")
    print("Energy Alert: No")

usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_res(usage)
```

**Problem Statement 6: Online Event Registration Capacity Controller**

```python
def manage_reg(capacity, registrations):
  confirmed = 0
  waitlist = 0
  for i in range(registrations):
    if confirmed < capacity:
      confirmed += 1
    else:
      waitlist += 1
  print("Confirmed Registratios: ", confirmed)
  print("Waitlisted Users:", waitlist)
  if waitlist >0:
    print("Registration Status: Closed")
  else:
    print("Registration Status: Open")

manage_reg(100, 105)
```
