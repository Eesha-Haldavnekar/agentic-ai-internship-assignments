**Problem 1: Social Media – Post Engagement Analyzer**

```python
def analyze_post_engagement():
  n = int(input("Enter no. of posts: "))
  total_likes = 0;
  for i in range(n):   #loop for calculation
    likes = int(input(f"Enter likes for post{i+1}: "))
    total_likes += likes
  print(f"\nTotal likes: {total_likes}")
  if total_likes >=1000:
    print("Post Status: Viral Post")
  else:
    print("Post Status: Normal Engagement")
analyze_post_engagement()    #function call
```

**Problem 3: Agriculture – Rainfall Adequacy Checker**

```python
def rainfall_checker():
  n = int(input("Enter number of months: "))
  total_rainfall = 0
  for i in range(n):    #to calculate average
    rainfall = float(input(f"Enter rainfall for month {i+1}: "))
    total_rainfall += rainfall
  average_rainfall = total_rainfall / n
  required_level = float(input("Enter required rainfall level: "))
  print(f"\nAverage Rainfall: {int(average_rainfall)}")
  if average_rainfall >= required_level:
    print("Rainfall Status: Adequate Rainfall")
  else:
    print("Rainfall Status: Inadequate Rainfall")

rainfall_checker()
```

**Problem 4: Social Media – Duplicate Account Detection**

```python
def detect_duplicate_accounts():
  n = int(input("Enter number of usernames: "))
  usernames = []
  for i in range(n):
    name = input(f"Enter username {i+1}: ")
    usernames.append(name)
  unique_usernames = set(usernames)
  if len(usernames) != len(unique_usernames):
    print("\nDuplicate Accounts Found: Yes")
  else:
    print("\nDuplicate Accounts Found: No")

detect_duplicate_accounts()
```

**Problem 6: Agriculture – Premium Crop Price Filter**

```python
def filter_premium_crops():
  n = int(input("Enter number of crops: "))
  prices = []
  for i in range(n):
    price = int(input(f"Enter price of crop {i+1}: ₹"))
    prices.append(price)
  premium_crops = []
  for price in prices:
    if price > 2000:
      premium_crops.append(price)
  print(f"\nPremium Crops: {premium_crops}")

filter_premium_crops()
```

**Problem 9: E-Learning – Student Attendance Eligibility System**

```python
def check_attendance_eligibility():
  total_classes = int(input("Enter total number of classes: "))
  attendance_list = []
  for i in range(total_classes):
    status = int(input(f"Enter attendance for class {i+1} (1 for Present, 0 for Absent): "))
    attendance_list.append(status)
  attended = 0
  for value in attendance_list:
    attended +=value
  attendance_percentage = (attended / total_classes) * 100
  print(f"\nAttendance Percentage: {attendance_percentage}")
  if attendance_percentage >= 75:
    print("Exam Eligibility: Eligible")
  else:
    print("Exam Eligibility: Not Eligible")

check_attendance_eligibility()
```

**Problem 10: Smart Electricity Bill Analyzer**

```python
def calculate_electricity_bill(units):
  bill = 0
  if units <=100:               #biling slabs
    bill = units*3
  elif units <= 200:
    bill = (100 * 3) + (units - 100) * 5
  else:
    bill = (100 * 3) + (100 * 5) + (units - 200) * 7

  if bill < 500:                 #usage classification
    status = "Low Usage"
  elif bill <= 1500:
    status = "Moderate Usage"
  else:
    status = "High Usage"
  return bill, status

units = int(input("Enter units consumed: "))
total_bill, usage_status = calculate_electricity_bill(units)
print(f"\nTotal Electricity Bill: ₹{total_bill}")
print(f"Usage Classification: {usage_status}")
```
