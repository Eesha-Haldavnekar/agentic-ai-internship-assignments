**Problem 1: Smart Light Controller**

```python
class Smartlight:
  def __init__(self, name):
    self.name = name
    self.status = "OFF"   #default value
  def turn_on(self):
    self.status = "ON"
    print(f"{self.name} is {self.status}")
  def turn_off(self):
    self.status = "OFF"
    print(f"{self.name} is {self.status}")

light_name = input("Light name: ")
action = input("Action (ON/OFF): ")

lgt = Smartlight(light_name)   #object
if action == "ON":
  lgt.turn_on()
elif action == "OFF":
  lgt.turn_off()
else:
  print("Invalid action")
```

**Problem 2: Employee ID Card System**

```python
class Employee:            #class creation
  def __init__(self, name, emp_id, dept):
    self.name = name
    self.emp_id = emp_id
    self.dept = dept
  def display_id_card(self):
    print(f"\nEmployee ID Card")
    print("Name:", self.name)
    print("ID:", self.emp_id)
    print("Department:", self.dept)

name = input("Employee Name: ")
emp_id = input("Employee ID: ")
dept = input("Department: ")

emp = Employee(name, emp_id, dept)   #object
emp.display_id_card()
```

**Problem 4: Product Price Tag Generator**

```python
class Product:
  def __init__(self, name, price):
    self.name = name
    self.price = price
  def display_price_tag(self):
    print(f"\nProduct:", self.name)
    print("Price: ₹", self.price)

name = input("Product name: ")
price = input("Price: ")

pr = Product(name, price) #object
pr.display_price_tag()
```

**Problem 5: Movie Rating Display System**

```python
class Movie:
  def __init__(self, name, rating):
    self.name = name
    self.rating = rating
  def display_rating(self):
    print(f"\nMovie:", self.name)
    print("Rating:", self.rating, "/5")

name = input("Movie name: ")
rating = input("Rating: ")

m1 = Movie(name, rating)
m1.display_rating()
```
