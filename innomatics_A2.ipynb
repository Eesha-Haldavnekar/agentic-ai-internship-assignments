Problem 1: Unique Words in a Sentence


```python
sentence = input("Enter a sentence: ")
sentence = sentence.lower()
words = sentence.split()
unique_words = set(words)
print("Unique words count: ", len(unique_words))
print("Unique words: ", unique_words)
```

Problem 2: Highest Salary from Employee Data

```python
#dictionary
employees = {
      "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
highest_paid_emp = max(employees, key = employees.get)
highest_salary = employees[highest_paid_emp]
print("Highest Salary: ", highest_paid_emp, "-", highest_salary)
```

Problem 3: Find Maximum and Minimum Values


```python
numbers = [45, 22, 89, 10, 66]
max_val = numbers[0]
min_val = numbers[0]
for num in numbers:
  if num > max_val:
    max_val = num
  if num < min_val:
    min_val = num
print("List: ", numbers)
print("Max: ", max_val)
print("Min: ", min_val)
```

Problem 4: Count Products Above a Price Threshold


```python
prizes = [450, 1200, 899, 1500, 300]
count = 0
for price in prizes:
  if price > 1000:
    count += 1
print("Products above 1000: ", count)
```

Problem 5: Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]
present_count = attendance.count("P")
attendance_percentage = (present_count / len(attendance))*100
print("Attendace Percentage: ", attendance_percentage)
```

Problem 7: Count Character Frequency


```python
text = "python"
char_count = {}
for char in text:
  if char in char_count:
    char_count[char] +=1
  else:
    char_count[char] = 1
print(char_count)
```
